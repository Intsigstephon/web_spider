```python
from bs4 import BeautifulSoup
sample = ' \
<html> \
<h1 id="title">hello world </h1> \
<a href="#" class="link" >this is link1</a> \
<a href="# link2" class="link" >this is link2</a> \
<body> \
</body> \
</html>'
```

```python
soup = BeautifulSoup(sample)
print(type(soup))
print(soup.text)
```

```python
#使用select筛选出h1标签的内容
h1 = soup.select('h1')
print(type(h1))
print(h1)
print(h1[0].text)
```

```python
#使用select筛选出所有包含a标签的内容
links = soup.select('a')
print(type(links))
print(links)
for link in links:
    print(link.text)
```

```python
#如果要查找特定属性的内容：
#比如说： id属性为#;  class属性为.
alink = soup.select('#title')
print(alink)

for link in soup.select('.link'):
    print(link)
```

```python
#使用select找出所有a里面的href
alink = soup.select("a")
for link in alink:
    print(link)
    print(link['href'])
```

```python
#注意区分，什么是标签，什么事属性。
#标签，含有属性。  标签： 比如说h1, h2, h3, a
#id属性用#， class属性用.  都可以用select方法进行获取
```
